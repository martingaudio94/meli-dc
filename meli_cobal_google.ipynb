```
entrada=Input((21,))
embedding=Embedding(maxim, 32)(entrada)



lstm=Bidirectional(GRU(512, dropout=0.1, recurrent_dropout=0.3,return_sequences=True))(embedding)
drop1=Dropout(.2)(lstm)
lstm3=GRU(1024, dropout=0.05, recurrent_dropout=0.35,return_sequences=False)(drop1)
drop2=Dropout(.2)(lstm3)
out=Dense(1588,activation='softmax')

salida=out(drop2)

model=Model(entrada,salida)
RMS=rmsprop(lr=0.001, rho=0.9, epsilon=None, decay=0.00001)

model.compile(loss = 'categorical_crossentropy', optimizer=RMS ,metrics = ['accuracy'])
```
