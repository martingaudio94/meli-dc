```
from keras.layers import Bidirectional,GRU
```

```
# este model se creo para el entrenamiento de la capa embedding
# al modelo se le quitan todas las capas menos la embedding y la capa de entrada
# luego se hace fine-tunning se freeza la capa y se utiliza en el modelo siguiente
```

```
entrada=Input((36,))
embedding=Embedding(maxim, 50)(entrada)



lstm=Bidirectional(GRU(512, dropout=0.1, recurrent_dropout=0.3,return_sequences=True))(embedding)
drop1=Dropout(.2)(lstm)
lstm3=GRU(1024, dropout=0.05, recurrent_dropout=0.35,return_sequences=True)(drop1)
drop2=Dropout(.2)(lstm3)
LSTM3=GRU(512,recurrent_dropout=.3)(lstm3)
frop3=Dropout(.2)(LSTM3)
out=Dense(1588,activation='softmax')

salida=out(frop3)

model=Model(entrada,salida)
RMS=rmsprop(lr=0.001, rho=0.9, epsilon=None, decay=0.00001)

model.compile(loss = 'categorical_crossentropy', optimizer=RMS ,metrics = ['accuracy'])
```

```

b=model.layers[2](model.layers[1].output)
c=model.layers[3](b)
d=model.layers[4](c)
j=model.layers[5](d)
k=model.layers[6](j)
e=model.layers[-1](k)
modelo=Model(model.layers[0].input,e)
rms=rmsprop(lr=0.001, rho=0.9, epsilon=None, decay=0.00001)
modelo.compile(loss = 'categorical_crossentropy', optimizer=RMS ,metrics = ['accuracy'])
```
